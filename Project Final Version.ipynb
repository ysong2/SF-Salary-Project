# Data Science Final Project

```python
import os
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Read data file

```python
df = pd.read_csv('data/sf2015.csv')
df
```

```python
new_df = df.drop(df.columns[[0,3,4,5,8,9,10]], axis=1) 
new_df
```

```python
T300k_jobs = new_df['Total Pay & Benefits'] > 300000
full_time = new_df['Status'] == 'FT'
over_300k= df[T300k_jobs & full_time][['Job Title','Total Pay & Benefits', 'Status']]
top_paid = over_300k[:31]
top_paid
```

# Group Top Paid Jobs by Job Titles

```python
datt = top_paid[['Job Title', 'Total Pay & Benefits']].groupby('Job Title').mean()
datt
sort_jobs = datt.sort(["Total Pay & Benefits"], ascending=False)
sort_jobs
```

# Make a Bokeh Graph

```python
from bokeh.charts import Bar, show, output_file, defaults
from bokeh.layouts import row

defaults.width = 1000
defaults.height= 1200


```

```python
bar = Bar(sort_jobs, title="SF jobs over 300k",legend="top_right",
             xlabel='Total Pay & Benefits', ylabel='Job Titles')
output_file("bar.html", title="bar.py Job pay")
show(bar)

```

# Percentage of Base Pay over Total Pay & Benefits 

```python
base_pay= new_df['Base Pay']>0
total = new_df['Total Pay & Benefits']
full_time = new_df['Status'] == 'FT'
new = new_df[base_pay & full_time & total][['Base Pay','Total Pay & Benefits', 'Status']]
new
```

```python
numerator = new['Base Pay'].dropna()
denominator = total.dropna()

# calculate percent BasePay of TotalPayBenefits
base_over_total = numerator / denominator
base_over_total
```

```python
fig, ax = plt.subplots(figsize=(8, 8))


sns.kdeplot(numerator, label="Total Pay & Benefit", shade=True, ax=ax)
sns.kdeplot(denominator, label="Base Pay", shade=True, ax=ax)

title = plt.title('Base salary and Total Pay & Benefit Distribution')
```
